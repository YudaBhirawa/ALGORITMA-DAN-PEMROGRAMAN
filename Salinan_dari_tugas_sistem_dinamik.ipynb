```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
X0 = 1
t = np.linspace(0, 5, 200)
ks = [1, 0, -1]
labels = ["k > 0 (1)", "k = 0", "k < 0 (-1)"]

plt.figure()
for k, label in zip(ks, labels):
    x = X0 * np.exp(k * t)
    plt.plot(t, x, label=label)

plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Solusi x(t) = X0 e^{kt} untuk berbagai nilai k")
plt.legend()
plt.grid(True)
plt.show()
```
